```
from keras.datasets import mnist
import seaborn as sns
# the data, shuffled and split between a train and test sets 
(X_train, y_train), (X_test, y_test) = mnist.load_data()
print('before reshape the data',X_train.shape)
X_train=X_train.reshape(X_train.shape[0],X_train.shape[1]*X_train.shape[2])
X_test=X_test.reshape(X_test.shape[0],X_test.shape[1]*X_test.shape[2])
print('After reshape the data',X_train.shape)

print('X_train',X_train.shape)
print('X_test',X_test.shape)
```

**Reshape the data**

```
X_train = X_train/255.0
X_test = X_test/255.0
```

**Print the label for first image and after converting it to categorical**

```
import keras
print("class lebel of first image  :", y_train[0])
num_classes = 10
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)

print("After converting class lebel of first image  :", y_train[0])
```


**Create a single layer perceptron with**
Single dense 1024 neurons

input shape being 784 activation fucntion relu.

Softmax layer with 10 neuron as output

loss- categorical_crossentropy

optimizer - RMSPROP

batch_size = 5000, epochs=50

Print the accuracy and loss of SLP model






```
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import adam
from keras import optimizers

from keras.utils.vis_utils import plot_model
batch_size = 5000
num_classes = 10
epochs = 50
model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(784,)))
model.add(Dense(num_classes, activation='softmax'))

model.summary()

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.fit(X_train, y_train,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_data=(X_test, y_test))
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

**Print the accuracy**

```
loss, acc = model.evaluate(X_test, y_test, verbose=0)
print('Test loss:', loss)
print('Test acc:', acc)
```

#**Multilayer Perceptron**

Single dense layer 1024 neurons

input shape being 784

activation fucntion relu.

Second Dense Layer 512 neurons and activation relu

Thrid Dense Layer 512 Neurons and activaton relu

Softmax layer with 10 neuron as output

loss- categorical_crossentropy

optimizer - adam

batch_size = 5000, epochs=50

Print the accuracy and loss of MLP model

```
from keras.layers import Activation, Dense

model_mlp = Sequential()
model_mlp.add(Dense(1024, input_shape=(784, )))
model_mlp.add(Activation('relu'))
model_mlp.add(Dense(512))
model_mlp.add(Activation('relu'))
model_mlp.add(Dense(512))
model_mlp.add(Activation('relu'))
model_mlp.add(Dense(10))
model_mlp.add(Activation('softmax'))

model_mlp.compile(loss='categorical_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])
model_mlp.fit(X_train, y_train,
                        batch_size=5000,
                        nb_epoch=50,
                        verbose=1,
                        validation_data=(X_test, y_test))
loss, acc = model_mlp.evaluate(X_test, y_test, verbose=0)
print('Test loss:', loss)
print('Test acc:', acc)
plot_model(model_mlp, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

# **Convolutional neural network**


- Conv2D with 32 Neuron; Filter 3,3 ; Activation: Relu ; Stride (1,1) 
- MaxPool2D ; Pool Size (2,2) 
- Flatten the data again to send to desne layer 
- 128 Neuron single Dense Layer with relu
- 10 neuron single dense layer with softmax as output layer

```
(X_train, y_train), (X_test, y_test) = mnist.load_data()
print('before reshape the data',X_train[1].shape)
```

```
X_train = X_train.reshape((60000, 28, 28, 1))
X_test = X_test.reshape((10000, 28, 28, 1))

X_train = X_train/255.0
X_test = X_test/255.0

print('After reshape the data',X_train.shape)
```

```
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import  Conv2D ,MaxPool2D
from keras.layers import Activation,Dropout,Flatten, Dense
from keras import backend as k
import numpy as np
from keras.preprocessing import image
import tensorflow as tf
model = Sequential()
model.add(Conv2D(32,(3,3),input_shape=(28,28,1)))
model.add(Activation('relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dense(10))
model.add(Activation('softmax'))


model.summary()

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.fit(X_train, y_train,
                        batch_size=5000,
                        nb_epoch=50,
                        verbose=1,
                        validation_data=(X_test, y_test))
loss, acc = model.evaluate(X_test, y_test, verbose=0)
print('Test loss:', loss)
print('Test acc:', acc)
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

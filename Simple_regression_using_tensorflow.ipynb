```
m =tf.Variable(0.78)

b= tf.Variable(0.56)
```

```
error = 0

init = tf.global_variables_initializer()

sess=tf.InteractiveSession()
sess.run(init)
for x,y in zip(x_data,y_lbl):
  #predicted y value
  #print(x)
  #print(b)

  y_hat = m*x+b

  error +=(y-y_hat)**2  #cost function
  print("actual value",y)
  print("predicted value",y_hat)
  print("error value",error)
  print('\n')
```

```
optimizer= tf.train.GradientDescentOptimizer(learning_rate=0.001)


train = optimizer.minimize(error)

#print(train)

training_step=1000

for i in range(training_step):
  sess.run(train)
  print('train')

final_slope,final_intercept = sess.run([m,b])

print(final_slope)

print(final_intercept)
```

```
x_test = np.linspace(-2,11,10)

#y = mx+b

y_predplot = final_slope*x_test+final_intercept

print(y_predplot)

plt.plot(x_test,y_predplot,'r')


```
